# Data Transformation

Data Transformation จาก booking data ให้เป็น stay data เนื่องจาก ในธุรกิจโรงแรม ตัวชี้วัดสำคัญ (KPIs) เช่น Occupancy Rate, Average Daily Rate (ADR) และ Revenue per Available Room (RevPAR) ล้วนถูกคำนวณบนพื้นฐานของ รายวัน หรือที่เรียกว่า daily SKU ซึ่งเป็นหน่วยที่เล็กที่สุดของการให้บริการ

ข้อมูลในรูปแบบ Booking Data (1 แถวต่อ 1 การจอง) ยังไม่สอดคล้องกับลักษณะของ KPIs ดังกล่าว เนื่องจาก 1 การจองสามารถครอบคลุมหลายวันเข้าพัก ทำให้ไม่สามารถนำไปคำนวณตัวชี้วัดรายวันได้โดยตรงอย่างถูกต้อง

ดังนั้น จึงจำเป็นต้องทำ Data Transformation โดยแปลงข้อมูลจากระดับการจอง (booking-level) ให้เป็นระดับการเข้าพักรายวัน (stay-level หรือ daily grain) ซึ่งมีลักษณะเป็น 1 แถวต่อ 1 วันต่อ 1 การเข้าพัก (room-night) เพื่อให้สอดคล้องกับหน่วยของการวัดผลทางธุรกิจ

การปรับ Data Granularity ให้อยู่ในระดับที่เล็กที่สุด (lowest granularity) จะช่วยให้:

* สามารถคำนวณ KPIs รายวันได้อย่างถูกต้องและแม่นยำ
* รองรับการวิเคราะห์ในมิติต่าง ๆ เช่น รายสัปดาห์ รายเดือน หรือรายฤดูกาล ได้อย่างยืดหยุ่นผ่านการ aggregate ข้อมูล
* ลดความซับซ้อนในการคำนวณ และป้องกันความคลาดเคลื่อนจากการกระจายข้อมูลย้อนหลัง
* สอดคล้องกับลักษณะการดำเนินธุรกิจโรงแรมที่มีการขายห้องพักในรูปแบบ ต่อคืน


```python
df = pd.read_csv("/content/fact_bookings.csv")

# แปลงเป็น datetime
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['check_out_date'] = pd.to_datetime(df['check_out_date'])

# สร้าง list ของวันที่
df['stay_date'] = df.apply(
    lambda row: pd.date_range(
        start=row['check_in_date'],
        end=row['check_out_date'] - pd.Timedelta(days=1)
    ),
    axis=1
)

df_stay = df.explode('stay_date')
```

```python
df_stay.to_csv('stay_data.csv', index=False)
```

```python
print("Number of rows & columns")
print(f"Rows: {df_stay.shape[0]}")
print(f"Columns: {df_stay.shape[1]}")
```
